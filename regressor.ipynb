```python
import numpy as np

perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0,
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5,
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5,
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0,
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0,
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
import matplotlib.pyplot as plt

plt.scatter(perch_length,perch_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
print(len(perch_length))
print(len(perch_weight))
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)

print(train_input.shape, test_input.shape)
```

```python
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)

print(train_input.shape, test_input.shape)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=5) #5가 default 값임 
knr.fit(train_input, train_target)

print(knr.score(test_input, test_target)) # 0.992809406101064
```

```python
# train_input 점수와 비교해보기
print(knr.score(train_input, train_target)) # 0.9698823289099254 -> 과소적합
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=3) #5가 default 값임 
knr.fit(train_input, train_target)

print(knr.score(test_input, test_target))

print(knr.score(train_input, train_target))
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=1) #5가 default 값임 
knr.fit(train_input, train_target)

print(knr.score(test_input, test_target))

print(knr.score(train_input, train_target))
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=7) #5가 default 값임 
knr.fit(train_input, train_target)

print(knr.score(test_input, test_target))

print(knr.score(train_input, train_target))
```

```python
### 수동 예측
t_input = test_input[2]
p = knr.predict(t_input.reshape(1,-1))

print(f"예측 사용 무게 : {p}")
print(f"실 무게:{test_target[2]}")
```

```python
## 그래프로 그려보기
knr = KNeighborsRegressor()
x = np.arange(5,45).reshape(-1,1)

for k in [1,3,5,7,10]:
    knr.n_neighbors = k
    knr.fit(train_input, train_target)
    prediction = knr.predict(x)

    #결정R^2
    s_train = knr.score(train_input, train_target)
    s_test = knr.score(test_input, test_target)
    
    plt.title(f"k={k}, R^2={s_train-s_test}")
    plt.scatter(train_input, train_target)
    plt.plot(x, prediction)
    plt.xlabel('length')
    plt.ylabel('weight')
    plt.show()
```

```python
####### k=3이 가장 좋다 (?) 왜
knr = KNeighborsRegressor()
knr.n_neighbors = 3
knr.fit(train_input, train_target)
print(knr.predict([[20]]))
print(knr.predict([[34]]))
print(knr.predict([[35]]))
print(knr.predict([[36]]))
print(knr.predict([[37]]))
print(knr.predict([[42]])) # 여기서 부터 예측 값이 같음 - 예측 범위를 갖고있음/ 다른 알고리즘이 필요하다.
print(knr.predict([[43]]))
print(knr.predict([[48]]))
print(knr.predict([[49]]))
print(knr.predict([[50]]))
print(knr.predict([[100]]))  # 일정 수준 이상 크기의 물고기는 KNN으로 예측할 수 없다. 
```

```python
# 이웃을 찾아 차트로 그리기
w=50
d,i = knr.kneighbors([[w]])

#훈련 데이터의 산점도 가가
plt.scatter(train_input, train_target)
plt.scatter(train_input[i], train_target[i], marker='D') #이웃 3
plt.scatter(50,1033.33333333, marker='*') # 50 이상인 값
plt.scatter(100,knr.predict([[100]])[0], marker='^')
plt.scatter(w,knr.predict([[w]])[0], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
